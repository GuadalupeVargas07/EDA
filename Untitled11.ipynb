```python
# Función de búsqueda lineal
def busqueda_lineal(lista, valor):
    for i, elemento in enumerate(lista):
        if elemento == valor:
            return i # Retorna la posición en la lista
    return -1 # Si no se encuentra el valor

# Función para mostrar el menú
def mostrar_menu():
    print("\nMenú de Opciones:")
    print("1. Agregar elemento a la lista")
    print("2. Realizar búsqueda lineal")
    print("3. Mostrar lista")
    print("4. Salir")

# Función principal
def programa_principal():
    lista = [] # Lista vacía para almacenar elementos
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            elemento = input("Ingrese el elemento a agregar: ")
            lista.append(elemento)
            print(f"Elemento '{elemento}' agregado.")

        elif opcion == "2":
            valor = input("Ingrese el valor a buscar: ")
            posicion = busqueda_lineal(lista, valor)
            if posicion != -1:
                print(f"El valor '{valor}' se encontró en la posición {posicion}.")
            else:
                print(f"El valor '{valor}' no se encuentra en la lista.")

        elif opcion == "3":
            print("Lista actual:", lista)

        elif opcion == "4":
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecutar el programa principal
programa_principal()
```
