```python
# Definición de la clase Restaurante para almacenar detalles
class Restaurante:
    def __init__(self, nombre, direccion, calificacion, especialidad):
        self.nombre = nombre
        self.direccion = direccion
        self.calificacion = calificacion
        self.especialidad = especialidad

    def __str__(self):
        return f"{self.nombre}, Dirección: {self.direccion}, Calificación: {self.calificacion}, Especialidad: {self.especialidad}"

# Lista predefinida de restaurantes
lista_restaurantes = [
    Restaurante("Gorditas Gabino", "Calle Constitución 112", "4.5/5", "Mexicana, conocida por sus gorditas"),
    Restaurante("Mendoza Restaurante Galería", "Calle Constitución 200", "4.4/5", "Mexicana con ambiente de galería"),
    Restaurante("Lucky Black Pizza Pasta y Vino", "Calle Fany 241", "4.5/5", "Italiana, especializada en pizzas y pasta"),
    Restaurante("Lugano", "Calle Pino Suárez 350", "4.6/5", "Italiana con un ambiente acogedor"),
    Restaurante("Esquilón", "Av. 20 de Noviembre 120", "4.4/5", "Mexicana e internacional"),
    Restaurante("La Victoria", "Calle Negrete 181", "4.3/5", "Mexicana, famosa por sus platillos tradicionales"),
    Restaurante("Don Quijote Café & Cucina", "Calle Juárez 76", "4.2/5", "Mexicana y cocina café"),
    Restaurante("Barocco", "Av. Hidalgo 156", "4.3/5", "Italiana y mediterránea"),
    Restaurante("El Zocabon", "Av. 5 de Febrero 456", "4.3/5", "Mexicana tradicional"),
    Restaurante("Santa Fortuna Bakery & Café", "Calle Independencia 240", "4.2/5", "Mexicana y cafetería"),
    Restaurante("La Fogata", "Negrete 181", "4.0/5", "Churrasquería especializada en cortes de carne"),
    Restaurante("Pan Bendito", "Av. México 235", "4.1/5", "Mexicana y panadería"),
    Restaurante("Corleone", "Blvd. Dolores del Río 305", "4.0/5", "Italiana y pizzería"),
    Restaurante("Wallander", "Calle Victoria 124", "4.5/5", "Café y comida ligera"),
    Restaurante("Cafe Central", "Av. 20 de Noviembre 118", "4.2/5", "Café y postres"),
    Restaurante("Naturalia", "Av. Reforma 322", "4.1/5", "Comida saludable"),
    Restaurante("Bliss Café", "Calle Juárez 88", "4.2/5", "Mexicana y café"),
    Restaurante("Restaurant La Finca", "Av. 20 de Noviembre 100", "4.0/5", "Mexicana e internacional"),
    Restaurante("Taquito Grill Tacos & Sport Bar", "Av. Constitución 234", "4.1/5", "Mexicana y bar"),
    Restaurante("Wings Army", "Calle Juárez 152", "4.0/5", "Bar y pub")
]

# Función de búsqueda lineal según el criterio
def busqueda_lineal(lista, valor, criterio):
    resultados = []
    for restaurante in lista:
        if criterio == "nombre" and valor.lower() in restaurante.nombre.lower():
            resultados.append(restaurante)
        elif criterio == "direccion" and valor.lower() in restaurante.direccion.lower():
            resultados.append(restaurante)
        elif criterio == "calificacion" and valor.lower() == restaurante.calificacion.lower():
            resultados.append(restaurante)
        elif criterio == "especialidad" and valor.lower() in restaurante.especialidad.lower():
            resultados.append(restaurante)
    return resultados

# Función para mostrar el menú
def mostrar_menu():
    print("\nMenú de Opciones:")
    print("1. Agregar nuevo restaurante")
    print("2. Buscar restaurante específico")
    print("3. Realizar búsqueda por criterio")
    print("4. Mostrar lista de restaurantes")
    print("5. Salir")

# Función principal
def programa_principal():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            # Recolección de datos para el nuevo restaurante
            nombre = input("Ingrese el nombre del restaurante: ")
            direccion = input("Ingrese la dirección del restaurante: ")
            calificacion = input("Ingrese la calificación del restaurante: ")
            especialidad = input("Ingrese la especialidad del restaurante: ")

            # Crear instancia de Restaurante y agregarla a la lista
            nuevo_restaurante = Restaurante(nombre, direccion, calificacion, especialidad)
            lista_restaurantes.append(nuevo_restaurante)
            print(f"Restaurante '{nombre}' agregado.")

        elif opcion == "2":
            # Búsqueda de un restaurante específico
            nombre_busqueda = input("Ingrese el nombre del restaurante que busca: ")
            resultado = busqueda_lineal(lista_restaurantes, nombre_busqueda, "nombre")
            if resultado:
                print("\nRestaurante encontrado:")
                for restaurante in resultado:
                    print(restaurante)
            else:
                print(f"No se encontró el restaurante '{nombre_busqueda}'.")

        elif opcion == "3":
            # Selección del criterio de búsqueda
            print("\nBuscar restaurante por:")
            print("1. Nombre")
            print("2. Dirección")
            print("3. Calificación")
            print("4. Especialidad")
            criterio_opcion = input("Seleccione el criterio de búsqueda: ")

            # Asignar criterio en función de la elección
            criterio = ""
            if criterio_opcion == "1":
                criterio = "nombre"
            elif criterio_opcion == "2":
                criterio = "direccion"
            elif criterio_opcion == "3":
                criterio = "calificacion"
            elif criterio_opcion == "4":
                criterio = "especialidad"
            else:
                print("Criterio no válido.")
                continue

            valor_busqueda = input(f"Ingrese el {criterio} del restaurante a buscar: ")
            resultados = busqueda_lineal(lista_restaurantes, valor_busqueda, criterio)
            if resultados:
                print("\nRestaurantes encontrados:")
                for restaurante in resultados:
                    print(restaurante)
            else:
                print(f"No se encontraron restaurantes con el {criterio} '{valor_busqueda}'.")

        elif opcion == "4":
            # Mostrar todos los restaurantes
            print("\nLista de restaurantes:")
            for restaurante in lista_restaurantes:
                print(restaurante)

        elif opcion == "5":
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecutar el programa principal
programa_principal()
```
