```python
# Instalar graphviz si no está instalado
!pip install graphviz

from graphviz import Digraph

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None
        self.derecha = None

class ArbolBinario:
    def __init__(self):
        self.raiz = None

    def insertar(self, valor):
        if not self.buscar(valor):  # Verifica si el valor ya existe antes de insertar
            self.raiz = self._insertar_recursivo(self.raiz, valor)
            print(f"Valor {valor} insertado en el árbol.")
        else:
            print(f"Valor {valor} ya existe en el árbol, no se insertó.")

    def _insertar_recursivo(self, nodo, valor):
        if nodo is None:
            return Nodo(valor)

        if valor < nodo.valor:
            nodo.izquierda = self._insertar_recursivo(nodo.izquierda, valor)
        else:
            nodo.derecha = self._insertar_recursivo(nodo.derecha, valor)

        return nodo

    def buscar(self, valor):
        return self._buscar_recursivo(self.raiz, valor)

    def _buscar_recursivo(self, nodo, valor):
        if nodo is None:
            return False
        if nodo.valor == valor:
            return True
        elif valor < nodo.valor:
            return self._buscar_recursivo(nodo.izquierda, valor)
        else:
            return self._buscar_recursivo(nodo.derecha, valor)

    def visualizar(self):
        dot = Digraph()
        dot.attr(rankdir='TB')
        self._visualizar_recursivo(self.raiz, dot)
        return dot

    def _visualizar_recursivo(self, nodo, dot):
        if nodo:
            dot.node(str(id(nodo)), str(nodo.valor))
            if nodo.izquierda:
                dot.edge(str(id(nodo)), str(id(nodo.izquierda)))
                self._visualizar_recursivo(nodo.izquierda, dot)
            if nodo.derecha:
                dot.edge(str(id(nodo)), str(id(nodo.derecha)))
                self._visualizar_recursivo(nodo.derecha, dot)

# Función para mostrar el menú
def mostrar_menu():
    print("\nMenú de Opciones:")
    print("1. Insertar valor en el árbol")
    print("2. Buscar valor en el árbol")
    print("3. Visualizar el árbol")
    print("4. Salir")

# Ejemplo de uso con menú
arbol = ArbolBinario()

while True:
    mostrar_menu()
    opcion = input("Selecciona una opción: ")

    if opcion == '1':
        valor_insertar = input("Ingresa el valor que deseas insertar: ")
        try:
            arbol.insertar(int(valor_insertar))
        except ValueError:
            print("Por favor, ingresa un número válido.")

    elif opcion == '2':
        valor_buscar = input("Ingresa el valor que deseas buscar: ")
        try:
            encontrado = arbol.buscar(int(valor_buscar))
            print(f"Valor {valor_buscar} {'encontrado' if encontrado else 'no encontrado'} en el árbol.")
        except ValueError:
            print("Por favor, ingresa un número válido.")

    elif opcion == '3':
        dot = arbol.visualizar()
        dot.render('arbol_binario', view=True, format='png')
        print("El árbol ha sido visualizado y guardado como 'arbol_binario.png'.")

    elif opcion == '4':
        print("Saliendo del programa.")
        break

    else:
        print("Opción no válida. Por favor, selecciona una opción del menú.")
```
