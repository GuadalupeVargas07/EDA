```python
import random

def generar_numeros(cantidad):
    """Generar una lista de números aleatorios entre 0 y 999."""
    return [random.randint(0, 999) for _ in range(cantidad)]

def imprimir_lista(lista, mensaje):
    """Imprimir una lista con un mensaje dado."""
    print(f"{mensaje}:")
    print(lista)

def burbuja_sort(lista):
    """Ordenar una lista usando el algoritmo de burbuja."""
    n = len(lista)
    for i in range(n):
        for j in range(0, n-i-1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return lista

def merge_sort(lista):
    """Ordenar una lista usando el algoritmo MergeSort."""
    if len(lista) > 1:
        mid = len(lista) // 2
        izquierda = lista[:mid]
        derecha = lista[mid:]

        merge_sort(izquierda)
        merge_sort(derecha)

        i = j = k = 0

        while i < len(izquierda) and j < len(derecha):
            if izquierda[i] < derecha[j]:
                lista[k] = izquierda[i]
                i += 1
            else:
                lista[k] = derecha[j]
                j += 1
            k += 1

        while i < len(izquierda):
            lista[k] = izquierda[i]
            i += 1
            k += 1

        while j < len(derecha):
            lista[k] = derecha[j]
            j += 1
            k += 1
    return lista

def main():
    cantidad = 1000
    numeros = generar_numeros(cantidad)
    imprimir_lista(numeros, "Lista desordenada")



    mitad = cantidad // 2
    vector1 = numeros[:mitad]
    vector2 = numeros[mitad:]

    print("------------------------------------------------------------------")

    imprimir_lista(vector1, "Primera mitad desordenada")
    imprimir_lista(vector2, "Segunda mitad desordenada")

    vector1_ordenado = burbuja_sort(vector1)
    vector2_ordenado = burbuja_sort(vector2)


    print("------------------------------------------------------------------")

    imprimir_lista(vector1_ordenado, "Primera mitad ordenada (Burbuja)")
    imprimir_lista(vector2_ordenado, "Segunda mitad ordenada (Burbuja)")

    print("------------------------------------------------------------------")

    numeros_ordenados_merge = merge_sort(numeros)
    imprimir_lista(numeros_ordenados_merge, "Lista ordenada (MergeSort)")

if __name__ == "__main__":
    main()

```
