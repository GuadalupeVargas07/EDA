```python
# Función de búsqueda binaria
def busqueda_binaria(lista, valor):
    inicio = 0
    fin = len(lista) - 1

    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == valor:
            return medio  # Retorna la posición en la lista
        elif lista[medio] < valor:
            inicio = medio + 1
        else:
            fin = medio - 1
    return -1  # Si no se encuentra el valor

# Función para mostrar el menú
def mostrar_menu():
    print("\nMenú de Opciones:")
    print("1. Agregar elemento a la lista")
    print("2. Realizar búsqueda binaria")
    print("3. Mostrar lista ordenada")
    print("4. Salir")

# Función principal
def programa_principal():
    lista = []  # Lista vacía para almacenar elementos
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            try:
                elemento = int(input("Ingrese el número a agregar: "))
                lista.append(elemento)
                lista.sort()  # Ordena la lista para que la búsqueda binaria funcione
                print(f"Elemento '{elemento}' agregado.")
            except ValueError:
                print("Por favor, ingrese un número válido.")

        elif opcion == "2":
            try:
                valor = int(input("Ingrese el número a buscar: "))
                posicion = busqueda_binaria(lista, valor)
                if posicion != -1:
                    print(f"El valor '{valor}' se encontró en la posición {posicion}.")
                else:
                    print(f"El valor '{valor}' no se encuentra en la lista.")
            except ValueError:
                print("Por favor, ingrese un número válido.")

        elif opcion == "3":
            print("Lista ordenada actual:", lista)

        elif opcion == "4":
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecutar el programa principal
programa_principal()1
```
