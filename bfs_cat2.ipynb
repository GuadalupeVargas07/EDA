```python
from collections import deque

class Graph:
    def __init__(self, size):
        self.adj_matrix = [[0] * size for _ in range(size)]
        self.size = size
        self.vertex_data = [''] * size

    def add_edge(self, u, v):
        if 0 <= u < self.size and 0 <= v < self.size:
            self.adj_matrix[u][v] = 1
            self.adj_matrix[v][u] = 1

    def add_vertex_data(self, vertex, data):
        if 0 <= vertex < self.size:
            self.vertex_data[vertex] = data

    def print_graph(self):
        print("Adjacency Matrix:")
        for row in self.adj_matrix:
            print(' '.join(map(str, row)))
        print("\nVertex Data:")
        for vertex, data in enumerate(self.vertex_data):
            print(f"Vertex {vertex}: {data}")

    def bfs_traverse(self, start_vertex_data):
        start_vertex = self.vertex_data.index(start_vertex_data)
        visited = [False] * self.size  # Para rastrear los vértices visitados
        queue = deque([start_vertex])  # Cola para el BFS
        visited[start_vertex] = True

        bfs_order = []  # Lista para almacenar el orden de recorrido

        print(f"BFS traversal starting from vertex {start_vertex_data}:")

        # BFS loop para recorrer todo el grafo
        while queue:
            current_vertex = queue.popleft()
            bfs_order.append(self.vertex_data[current_vertex])  # Agregar vértice al recorrido
            print(f"Visiting: {self.vertex_data[current_vertex]}")

            # Explorar todos los vecinos del vértice actual
            for i in range(self.size):
                if self.adj_matrix[current_vertex][i] == 1 and not visited[i]:
                    queue.append(i)
                    visited[i] = True
                    # Mostrar la relación (arista) entre los nodos
                    print(f"  {self.vertex_data[current_vertex]} -> {self.vertex_data[i]}")

        # Imprimir el recorrido final
        print("\nComplete BFS Traversal Order:")
        print(" -> ".join(bfs_order))
        print()  # Nueva línea al final del recorrido

# Creación del grafo con 11 vértices
g = Graph(11)

# Asignación de los datos de los vértices
g.add_vertex_data(0, 'A')
g.add_vertex_data(1, 'B')
g.add_vertex_data(2, 'C')
g.add_vertex_data(3, 'D')
g.add_vertex_data(4, 'E')
g.add_vertex_data(5, 'F')
g.add_vertex_data(6, 'G')
g.add_vertex_data(7, 'H')
g.add_vertex_data(8, 'I')
g.add_vertex_data(9, 'J')
g.add_vertex_data(10, 'K')

# Creación de las aristas del grafo
g.add_edge(0, 2)  # A-C
g.add_edge(0, 1)  # A-B
g.add_edge(0, 3)  # A-D
g.add_edge(1, 4)  # B-E
g.add_edge(1, 5)  # B-F
g.add_edge(3, 5)  # D-F
g.add_edge(3, 6)  # D-G
g.add_edge(5, 7)  # F-H
g.add_edge(5, 8)  # F-I
g.add_edge(8, 10) # I-K
g.add_edge(7, 9)  # H-J
g.add_edge(9, 10) # J-K

# Imprimir el grafo
g.print_graph()

# Realizar el recorrido BFS desde el vértice 'A'
print("\nBreadth First Search traversal from vertex A:")
g.bfs_traverse('A')
```
